```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt

# Normalized data
data = {
    'Oils/fats': ['Mustard oil', 'Desi ghee', 'Soya bean oil', 'Olive oil'],
    'Saturated fatty-acids': [0.00, 1.00, 0.18, 0.15],
    'Mono-unsaturated fatty acids': [0.88, 0.10, 0.00, 1.00],
    'Omega-6: omega-6 ratio': [0.00, 0.10, 0.50, 1.00]
}

# Create DataFrame
df = pd.DataFrame(data)

# Extract only the numerical columns for clustering
X = df[['Saturated fatty-acids', 'Mono-unsaturated fatty acids', 'Omega-6: omega-6 ratio']]

# Perform hierarchical clustering using complete linkage
Z = linkage(X, method='complete')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=df['Oils/fats'].values)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Oils/fats')
plt.ylabel('Distance')
plt.show()
```
